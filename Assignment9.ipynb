Before beginning the assignment, let me clarify this: Say we have two variables; x and y. From what I understand, a Pearson correlation between the two variables approaching 1 means that when x increases, y increases in a "straight" line also. A correlation approaching 0 means that the two variables have nothing do with eachother. A correlation approaching -1 means that when x increases, y decreases in a relatively "straight" line. Now, let's use the 'corr()' function to calculate all the correlations between the numerical data in our dataset.

I think that the higher the body mass, the higher the other numerical data. So data in column "body_mass_g", will have a positive Pearson correlation between itself and the other columns. 

```python
penguinsCorr = penguins.corr(numeric_only = True)
penguinsCorr.style.background_gradient(cmap = 'coolwarm').format(precision = 2)
```

It looks like the body mass of a penguin has a positive correlation with its flipper length and bill length. However, it has a negative correlation with bill depth.
