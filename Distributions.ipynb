# Different types of distributions with code

```python
import numpy as np
import matplotlib.pyplot as plt

# Uniform integers from 20 to 45
data = np.random.randint(20, 46, size=1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=26, edgecolor='black')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Normal distribution centered at 0
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=30, density=True, edgecolor='black')
plt.title('Normal Distribution (Bell Curve)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Skewed distribution: use exponential
data = np.random.exponential(scale=1, size=1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Skewed Distribution (Right-Skewed)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import binom

# 1000 experiments of flipping a coin 10 times
data = binom.rvs(n=10, p=0.5, size=1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=np.arange(-0.5,11.5,1), edgecolor='black')
plt.title('Binomial Distribution (Coin Flips)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import poisson

# Events happening with average 3 per interval
data = poisson.rvs(mu=3, size=1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=np.arange(-0.5,15.5,1), edgecolor='black')
plt.title('Poisson Distribution (Rare Events)')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```
