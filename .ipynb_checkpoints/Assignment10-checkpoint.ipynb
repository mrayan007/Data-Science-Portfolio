```python
import pandas as pd
import seaborn as sns
import scipy.stats as st
import numpy as np

heart = pd.read_csv('Datasets/heart.csv').dropna()
heart['HeartDisease'] = heart['HeartDisease'].map({0: 'No', 1: 'Yes'})
heart.head()
```

For this assignment, I have to perform a bivariate analysis on at least one combination of two columns in my dataset. I want to see the correlations between a patient's age and his Cholesterol level, and a patient's average resting heartbeat and his maximum resting heartbeat.

Now I'm no doctor but I assume that the higher the resting heartbeat and Cholesterol, the higher the chance of heart failure.

```python
heartCorr = heart.corr(numeric_only = True)
heartCorr.style.background_gradient(cmap = 'coolwarm').format(precision = 2)
```

I don't see a lot of heat in this table, so I'm concluding that there aren't a lot of positive correlations.
