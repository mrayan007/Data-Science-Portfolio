```python
import pandas as pd
import seaborn as sns

footballPlayers = pd.read_csv("top5_leagues_player.csv")
footballPlayers.head()
```

```python
categoricalData = footballPlayers.select_dtypes(include = 'object').columns.tolist()
print(f"The dataset has {len(categoricalData)} columns with categorical data.")
numericalData = footballPlayers.select_dtypes(include = 'number').columns.tolist()
print(f"The dataset has {len(numericalData)} columns with numerical data.")
```

As we can see here, in this dataset we have 12 columns with categorical data and 6 columns with numerical data. We are thus going to perform univariate analyses on the columns "nationality", "club" (categorical data), "height" and "price" (numerical data). 

First, I want to see what the most abundant nationality of players in this dataset is. For this, I'm going to visualize the statistics using a pie chart.

```python
footballPlayers['nationality'].value_counts().plot(kind = 'pie')
```

As messy as it looks, I think that the clear winner here is Spain.

Now, I want to see all the different clubs in my dataset. However, it is possible that the same club is listed multiple times, so for that I will be using the ".unique()" property to avoid the same club name appearing more than once.

```python
footballPlayers['club'].unique()
```

It looks like there are too many clubs for me to count by myself, so I'm going to count them using the "len()" function.

```python
clubAmount = len(footballPlayers['club'].unique())
print(f"There are {clubAmount} different clubs in my dataset.")
```

For the numerical data, let's begin with the "height" column. I want to see who the tallest player in my dataset is, for that, I'm using the ".max()" property.

```python
footballPlayers['height'].max()
```

Looks like the tallest player in my dataset is 206CM! But we don't know who that is.. let's fix that!

```python
maxHeight = footballPlayers['height'].max()
footballPlayers[footballPlayers['height'] == maxHeight]
```

Phew, now we know that his name is Isaak Touré.

Lastly, I want to check if there's any crazy difference in price between the players in my dataset, for that, I'm using a box plot visualization.

```python
footballPlayers['price'].plot(kind = 'box')
```
